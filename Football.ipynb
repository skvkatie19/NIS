```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
sns.set()

%config InlineBackend.figure_format = 'svg'
```

```python
df = pd.read_excel('football.xlsx')
```

```python
df.head()
```

```python
df.columns
```

```python
features = ['%ВД', 'ВД', '%ВГ', 'ВГ', 'ИД', 'ИГ']
```

```python
df[features].hist(figsize=(20, 30))
```

```python
russia = df[df['Страна'] == 'Россия']
spain = df[df['Страна'] == 'Испания']
germany = df[df['Страна'] == 'Германия']
italy = df[df['Страна'] == 'Италия']
france = df[df['Страна'] == 'Франция']
england = df[df['Страна'] == 'Англия']
```

```python
df[features].plot(kind='density', subplots=True, layout=(4, 2), 
                  sharex=False, figsize=(15, 10));
```
