```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
sns.set()

%config InlineBackend.figure_format = 'svg'
features = ['%ВД', 'ВД', '%ВГ', 'ВГ', 'ИД', 'ИГ']
```

Что интересного можно заметить в ящике с усами: ящик с усами показывает интервартильный размах распределения. 

- Минимальное доля выигрышей (дома/в гостях) = значению левого уса
- Вертикальная черта в ящике - медиана, у половина команд доля выигрышей (дома/в гостях) лежит до значения вертикальной черты, половина после
- Максимальная доля выигрышей (дома/в гостях) = значению правого уса
- И количество выбросов (outliners),аномальных наблюдений = кол-во точек за пределами усов.

Выборов должно быть мало, посмотрите какие это команды, когда будете выводить на слайды, могут спросить
